```python
#Part 5 : RDD Transformations
```

```python
#Using RDD operations, extract CustomerName and Email.
rdd_nom_email = rdd_clients.map(lambda row: (row.name, row.email))
rdd_nom_email.take(5)
```

```python
#Create an RDD containing only customers whose age is above 35.
rdd_plus_35 = rdd_clients.filter(lambda row: row.age > 35)
rdd_plus_35.take(5)
```

```python
#Using RDD transformations, extract email domains from customer emails.
rdd_domaines = rdd_clients.map(lambda row: row.email.split("@")[1])
rdd_domaines.take(5)
```

```python
#Count how many customers belong to each email domain using RDD operations.
rdd_compte_domaines = rdd_domaines.map(lambda domaine: (domaine, 1)).reduceByKey(lambda x, y: x + y)
rdd_compte_domaines.collect()
```

Q30. Identify the Logical Plan and Physical Plan from the explain output.
The df_joint.explain() output shows how Spark optimizes and runs your query. It is divided into two main parts:

1. The Logical Plan
It shows WHAT Spark will do, step-by-step, before looking at the hardware. In your terminal, look for:

Parsed & Analyzed Plans: Spark checks your code syntax and column names.

Optimized Logical Plan: Spark's optimizer automatically rearranges operations (like filtering data early) to make the query run much faster.

2. The Physical Plan
It shows HOW Spark will actually execute the query on the machine.

In your terminal, look for the == Physical Plan == section.

It lists low-level physical operations such as *BroadcastHashJoin (how it performs the join), *Project (how it selects columns), and *Filter.

Shortcut for your report: Just copy the text block under == Optimized Logical Plan == for your Logical Plan, and the text block under == Physical Plan == for your Physical Plan.
