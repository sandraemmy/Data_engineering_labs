```python
from pyspark.sql.functions import col, desc, lower
df = spark.read.csv("/Volumes/workspace/default/demo/Spotify data.csv", header=True, inferSchema=True)

df.printSchema()

df.show(5)
```

```python
#longest song released
from pyspark.sql.functions import desc
longest_song = df.orderBy(desc("duration_ms")).select("song_title", "artist", "duration_ms").limit(1)

longest_song.show(truncate=False)
```

```python
drake_count=df.filter(col("artist") == "Drake").count()
print("Nombre de chansons de Drake:", drake_count)
```

```python
love_songs=df.filter(lower(col("song_title")).contains("love")).select("song_title", "artist")
love_songs.show(20, truncate=False)
```

```python
number_song=df.filter(col("song_title").rlike("[0-9]")).select("song_title","artist")
number_song.show(20, truncate=False)
```

```python
bachelorette_playlist=df.orderBy(col("danceability").desc(),col("valence").desc()).select("song_title","artist","danceability","valence")
bachelorette_playlist.show(10,truncate=False)
```

```python
dance_battle=df.orderBy(col("energy").desc(),col("tempo").desc()).select("song_title","artist","energy")
dance_battle.show(truncate=False)
```

```python
top_artist=df.groupBy("artist").count().orderBy(desc("count"))
top_artist.show(10,truncate=False)
```

```python
breakup_playlist=df.orderBy(col("valence").asc(),col("energy").asc()).select("song_title", "artist", "valence", "energy")
breakup_playlist.show(10,truncate=False)
```
