```python
df = spark.table("customers")
df.show()
```

```python
df.filter(df.EMAIL.isNull()).show(truncate=False)
```

```python
df.filter(df.COUNTRY == "France").show()
```

```python
from pyspark.sql.functions import concat_ws
new_df = df.withColumn(
    "FULLNAME",
    concat_ws(
        " ",
        df.CONTACTFIRSTNAME,
        df.CONTACTLASTNAME
    )
)

new_df.show()
```

```python
group_df = df.groupBy("COUNTRY").count()

group_df.show()
```

```python
df.createOrReplaceTempView("customers_view")
spark.sql("""
SELECT COUNTRY, COUNT(*) AS TOTAL_CUSTOMERS
FROM customers_view
GROUP BY COUNTRY
""").show()
```
